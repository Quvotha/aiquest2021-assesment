```python
for c in ('name', 'description'):
    mapping = {v: i for i, v in enumerate(train[c].unique()) if len(v) > 0}
    train[c] = train[c].map(mapping)
```

```python
train['num_appearances_name'] = train.groupby('name')['name'].transform('count')
train['num_appearances_description'] = train.groupby('description')['description'].transform('count')
train
```

```python
agg_name = train.query('num_appearances_name > 1').groupby(['name', 'city', 'zipcode'])['y'].describe().sort_index()
agg_description = train.query('num_appearances_description > 1').groupby(['description', 'city', 'zipcode'])['y'].describe().sort_index()
# agg_both = train.groupby(['description', 'name', 'city', 'zipcode'])['y'].describe().sort_index()
```
